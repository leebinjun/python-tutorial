# Label & Button

```python
import tkinter as tk

window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x100')

var = tk.StringVar() # 声明字符串变量

# 标签对象
l = tk.Label(window, textvariable = var, bg = 'green', 
             font=('Arial', 12), width = 15, height = 2) 
l.pack() # 放置

on_hit = False # 起始全局变量
def hit_me():
    global on_hit # 声明全局变量
    if on_hit == False:
        on_hit = True
        var.set('you hit me')
    else:
        on_hit = False
        var.set('')

# 按钮对象
b = tk.Button(window, text='hit me', bg = 'pink',
               font=('Arial', 12), width = 15, height = 2, command = hit_me)
b.pack() # 放置

window.mainloop() # 循环显示
```

#  Listbox

```python
import tkinter as tk

window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

var1 = tk.StringVar() # 声明字符串变量

# 标签对象
l = tk.Label(window, textvariable = var1, bg = 'pink', 
             font=('Arial', 12), width = 8, height = 2) 
l.pack() # 放置

def point_selection():
    value = lb.get(lb.curselection()) # lb.curselection() 光标指示位
    var1.set(value)

# 按钮对象
b = tk.Button(window, text='point selsecion',
               width = 15, height = 2, command = point_selection)
b.pack() 

var2 = tk.StringVar()
var2.set((11,22,33,44))

# listbox
lb = tk.Listbox(window, list = var2)
lb.pack()

list_items = [1,2,3,4]
for item in list_items:
    lb.insert('end', item) # 'end' 方式插入
lb.insert(1, 'first') # 索引插入（位前，从0开始）
lb.insert(2, 'second')
lb.delete(0) # 索引删除（该位，从0开始）
    
window.mainloop() # 循环显示
```

# Radiobutton (单项选择按钮)

```python
import tkinter as tk

window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

var = tk.StringVar() # 声明字符串变量

# 标签对象
l = tk.Label(window, text = 'empty', bg = 'pink', 
             font=('Arial', 12), width = 20, height = 2) 
l.pack() # 放置

def print_selection():
    l.config(text = 'you have selected ' + var.get())
    # 修改标签 l 的参数 text ，var.get() 是字符串
    
# 单项选择按钮对象 把变量 var 赋值为 'A'
rb1 = tk.Radiobutton(window, text='Option A', variable = var, value = 'A',
                    command = print_selection)
rb1.pack()

rb2 = tk.Radiobutton(window, text='Option B', variable = var, value = 'B',
                    command = print_selection)
rb2.pack()

rb3 = tk.Radiobutton(window, text='Option C', variable = var, value = 'C',
                    command = print_selection)
rb3.pack()
    
window.mainloop() # 循环显示
```

# Scale (滑动条)

```python
import tkinter as tk

window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

# 标签对象
l = tk.Label(window, text = 'empty', bg = 'pink', 
             font=('Arial', 12), width = 20, height = 2) 
l.pack() # 放置

def print_selection(v):
    l.config(text = 'you have selected '+ v)
    # 修改标签 l 的参数 text ，var.get() 是字符串

# 滑动条对象 取值5-11，水平排布，不显值，标签单位长度2（5，7，9，11），保留两位小数0.01
# command 默认传入滑动条的值，长度length为像素值
s = tk.Scale(window, label = 'try me', from_=5, to=11, orient = tk.HORIZONTAL,
            length = 200, showvalue = 0, tickinterval = 2, resolution = 0.01,
            command = print_selection)
s.pack()
    
window.mainloop() # 循环显示
```

#  Checkbutton (多项选择按钮)

```python
import tkinter as tk

window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

# 标签对象
l = tk.Label(window, text = 'empty', bg = 'pink', 
             font=('Arial', 12), width = 20, height = 2) 
l.pack() # 放置

def print_selection():
    if (var1.get() == 0) & (var2.get() == 0):  # .get()
        l.config(text = 'I do not love either')
    elif (var1.get() == 0) & (var2.get() == 1):
        l.config(text = 'I love only C++')
    elif (var1.get() == 1) & (var2.get() == 0):
        l.config(text = 'I love only Python')
    else:
        l.config(text = 'I love both')

var1 = tk.IntVar() # 使用对象前必须声明类型  
var2 = tk.IntVar()

# VS 单项选择按钮对象 把变量 var 赋值为 'A'
# 多项选择按钮对象 选中则 var1 = 1 ，未选中则 var1 = 0 
cb1 = tk.Checkbutton(window, text = 'Python', variable = var1, onvalue = 1, offvalue = 0,
                    command = print_selection)
cb1.pack()
cb2 = tk.Checkbutton(window, text = 'C++', variable = var2, onvalue = 1, offvalue = 0,
                    command = print_selection)
cb2.pack()
    
window.mainloop() # 循环显示
```

#  Canvas

```python
import tkinter as tk

#window = tk.Toplevel() 
window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

canvas = tk.Canvas(window, bg = 'blue', height = 100, width = 200)
# 插入图片(只能gif格式)
image_file = tk.PhotoImage(file = 'pictures/ins.gif')
image = canvas.create_image(0, 0, anchor = 'nw', image = image_file) # anchor = 'NW' 锚定图片左上角到 (0,0)
# 画图
x0, y0, x1, y1 = 100, 50, 130, 80 # 设定坐标
line = canvas.create_line(x0, y0, x1, y1) # 创建直线
oval = canvas.create_oval(x0, y0, x1, y1, fill = 'red') # 创建圆(正方形内切创建)
arc = canvas.create_arc(x0, y0, x1, y1, start = 30, extent =90, fill = 'yellow') # 创建扇形，起始角度-旋转角度
rect = canvas.create_rectangle(x0, y0, x1, y1)
canvas.pack()

def move_it():
    canvas.move(rect, 4, 0) # 图层对象 rect 右移 4 ，下移 0
    canvas.scale(oval, int((x0 + x1)/2), int((y0 + y1)/2), 1.1, 1.1) # 图层对象 oval 以~为中心 水平缩放1.1，垂直缩放1.1
    canvas.scale(arc, int((x0 + x1)/2), int((y0 + y1)/2), 1.2, 1.2)

# 按钮move
b = tk.Button(window, text='move', bg = 'pink', command = move_it).pack()
                
window.mainloop() # 循环显示
```

#  Menubar (菜单)

```python
import tkinter as tk

#window = tk.Toplevel() 
window = tk.Tk() # 窗口对象（首字母大写）
window.title('my window')
window.geometry('200x200')

# 标签对象
tk.Label(window, text = '', bg = 'pink').pack() 

count = 0
def do_job():
    global count
    l.config(text = 'do' + str(count))
    count += 1

# 菜单对象 初始化->空的一条
menubar = tk.Menu(window)

file_menu = tk.Menu(menubar, tearoff = 0) # File 目录 master = menubar，不拆分
menubar.add_cascade(label = 'File', menu = file_menu)
file_menu.add_command(label = 'New', command = do_job) # File 目录下的 子命令
file_menu.add_command(label = 'Open', command = do_job)
file_menu.add_command(label = 'Save', command = do_job)
file_menu.add_separator() # 分割线
file_menu.add_command(label = 'Exit', command = window.quit)

edit_menu = tk.Menu(menubar, tearoff = 0) # Edit 目录 master = menubar，不拆分
menubar.add_cascade(label = 'Edit', menu = edit_menu)
edit_menu.add_command(label = 'Cut', command = do_job) # Edit 目录下的 子命令
edit_menu.add_command(label = 'Copy', command = do_job)
edit_menu.add_command(label = 'Paster', command = do_job)

sub_menu = tk.Menu(edit_menu) # Edit 目录下的 子目录 master = edit_menu，拆分
edit_menu.add_cascade(label = 'Submenu', menu = sub_menu) 
sub_menu.add_command(label = 'Import', command = do_job)

window.config(menu = menubar) # 更改窗口参数 菜单 menu
                
window.mainloop() # 循环显示
```

```python
import tkinter as tk 
from tkinter import messagebox
import pickle # 用于存放数据

window = tk.Tk()
window.title('Welcome to Next Innovation')
window.geometry('450x400')

# 画布对象 welcome image
canvas = tk.Canvas(window, height = 150, width = 450)
image_file = tk.PhotoImage(file = 'pictures/welcome.gif')
image = canvas.create_image(0, 0, anchor = 'nw', image = image_file)
canvas.pack(side = 'top')

# 标签对象 User name: & Password:
l_UserName = tk.Label(window, text = 'User name: ').place(x = 50, y =150)
l_Password = tk.Label(window, text = 'Password: ').place(x = 50, y =190)

var_UserName = tk.StringVar()
var_UserName.set('Next_Innovation@NI.com') # 默认用户名
var_Password = tk.StringVar()

# 输入框对象
e_UserName = tk.Entry(window, textvariable = var_UserName, width = 25).place(x = 160, y = 150)
e_Password = tk.Entry(window, textvariable = var_Password, width = 25, show = '*').place(x = 160, y = 190)

# 登录函数
def user_Login():
    user_name = var_UserName.get()
    user_password = var_Password.get()
    # 导入用户信息 users_info
    try:
        with open('files/users_info.pickle','rb') as users_file:
            users_info = pickle.load(users_file)
    except FileNotFoundError:
        with open('files/users_info.pickle','wb') as users_file:
            users_info = {'admin':'admin'} # 字典形式存储 users_info
            pickle.dump(users_info, users_file)
    # 判断登录信息是否匹配
    if user_name in users_info:
        if user_password == users_info[user_name]:
            messagebox.showinfo(title = 'Welcome', message = 'How are you ' + user_name + ' ?')
        else:
            messagebox.showerror(title = 'Error', message = 'your password is wrong, please try again!')
    else:
        is_SignUp = messagebox.askyesno(title = 'Welcome', 
                                        message = 'You have not sign up yet. Would you sign up today?')
        if is_SignUp:
            user_SignUp()

# 注册函数
def user_SignUp():
    # 登录确认函数
    def sign_up():
        user_name_n = var_UserName_n.get()
        user_password_n = var_Password_n.get()
        user_confirm_password_n = var_ConfirmPassword_n.get()
        # 载入用户信息用于比较
        with open('files/users_info.pickle','rb') as users_file:
            users_info = pickle.load(users_file)
        # 判断注册信息
        if user_password_n != user_confirm_password_n:
            messagebox.showerror(title = 'Error', message = 'Password and confirm password must be the same!')
        elif user_name_n in users_info:
            messagebox.showerror(title = 'Error', message = 'The user has already signed up!')
        else:
            users_info[user_name_n] = user_password_n
            with open('files/users_info.pickle','wb') as users_file:
                pickle.dump(users_info, users_file)
            messagebox.showinfo(title = 'Welcome', message = 'You have successfully signed up!')
            window_SignUp.destroy()
    
    # 弹出新窗口
    window_SignUp = tk.Toplevel(window)
    window_SignUp.title('Sign up window')
    window_SignUp.geometry('350x200')
    
    # 标签对象 User name: & Password:
    l_UserName_n = tk.Label(window_SignUp, text = 'User name: ').place(x = 10, y =10)
    l_Password_n = tk.Label(window_SignUp, text = 'Password: ').place(x = 10, y =40)
    l_Password_n = tk.Label(window_SignUp, text = 'Password: ').place(x = 10, y =70)
    
    var_UserName_n = tk.StringVar()
    var_UserName_n.set('Next_Innovation@NI.com') # 默认用户名
    var_Password_n = tk.StringVar()
    var_ConfirmPassword_n = tk.StringVar()

    # 输入框对象
    e_UserName_n = tk.Entry(window_SignUp, textvariable = var_UserName_n, width = 25).place(x = 100, y = 10)
    e_Password_n = tk.Entry(window_SignUp, textvariable = var_Password_n, width = 25, show = '*').place(x = 100, y = 40)
    e_ConfirmPassword_n = tk.Entry(window_SignUp, textvariable = var_ConfirmPassword_n, 
                                   width = 25, show = '*').place(x = 100, y = 70)
    
    # 按钮对象 Sign up
    b_SignUp_n = tk.Button(window_SignUp, text = 'Sign up', command = sign_up).place(x = 150, y = 100)
    
# 按钮对象 Login & Sign up
b_Login = tk.Button(window, text = 'Login', command = user_Login).place(x = 160, y = 250)
b_SignUp = tk.Button(window, text = 'Sign up', command = user_SignUp).place(x = 260, y = 250)

window.mainloop()
```
