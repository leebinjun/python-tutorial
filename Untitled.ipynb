```python
from pyecharts import options as opts
from pyecharts.charts import Geo, Page
from pyecharts.faker import Collector, Faker
from pyecharts.globals import ChartType, SymbolType

import pandas as pd

dict_all = {
    "全国大学生电子设计竞赛\n 中国大学生智能设计竞赛总决赛": "杭州",
    "中国机器人大赛暨RoboCup公开赛": "合肥",
    "国际大学生3D打印机设计邀请赛": "上海",
    "全国大学生嵌入式设计大赛": "大连",
    "下一代互联网技术创新大赛\n 国际大学生iCAN创新创业大赛": "北京",
    "国际大学生iCAN创新创业大赛": "成都",
    "大学生创新方法大赛\n 全国大学生创新创业年会": "黑龙江",
    "中国机器人及人工智能大赛": "广州",
    "中国高校机器人创意大赛": "余姚",
    "人工智能校园创新大赛": "南京",
}



df = pd.DataFrame(index=range(len(dict_all)))
df['destination'] = dict_all.values()
df['competition'] = dict_all.keys()


destination = df['destination']
competition = df['competition']

geo  = (
        Geo()
        .add_schema(
            maptype="china",
            is_roam=True,
            itemstyle_opts=opts.ItemStyleOpts(color="#D8D7D7", border_color="#111")
        )
        .add("destination", 
             [list(z) for z in zip(destination, competition)], 
             type_=ChartType.EFFECT_SCATTER,
             symbol_size=10,
             color='#708DF2',
            ) 
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Geo-competiton"),
        )
)

geo.render_notebook()
```
